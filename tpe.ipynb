# 2. Análisis preliminar



Con el ``.info()``, parecería no haber nulos. Pero después de analizar las columnas, observamos que en ``BANDA y/o RAMAL`` y ``PK`` hay "nulos" con el valor "-".

Lo vamos a remplazar por ``NaN`` para que sea identificado como un nulo

Creamos una copia del dataset para no pisar los datos originales. La vamos a llamar ``data_preprocessed``

```python
data_preprocessed = data.copy() 
data_preprocessed['PK'] = data_preprocessed['PK'].replace('-', pd.NA)
data_preprocessed['BANDA y/o RAMAL'] = data_preprocessed['BANDA y/o RAMAL'].replace('-', pd.NA)
data_preprocessed['PK'] = pd.to_numeric(data_preprocessed['PK']) 
data_preprocessed.info()
```

# **Hipotesis 1**

## Alta concentración de siniestros en ciertos tramos críticos en **AU 25 de Mayo**.
La siniestralidad en la AU 25 de Mayo no es uniforme, sino que se concentra significativamente en tres zonas críticas: 
- El *empalme inicial* **(PK 0-3000)**
- La zona de *salidas a barrios* **(PK 3000-6000)**
- El complejo de *bifurcación final* **(PK 6000-10000).**

Utilizamos la variable ``PK`` sobre la **autopista 25 de mayo**

```python
data_25mayo = data_preprocessed[data_preprocessed['AUTOPISTA'] == 'AU 25 DE MAYO']
data_25mayo = data_25mayo[data_25mayo['PK'].notna()] 
data_25mayo["PK"].describe()
```

```python
tramo_1 = data_25mayo[(data_25mayo['PK'] >= 0) & (data_25mayo['PK'] < 3000)]
tramo_2 = data_25mayo[(data_25mayo['PK'] >= 3000) & (data_25mayo['PK'] < 6000)]
tramo_3 = data_25mayo[(data_25mayo['PK'] >= 6000) & (data_25mayo['PK'] <= 10000)]

count_tramo_1 = len(tramo_1) 
count_tramo_2 = len(tramo_2)
count_tramo_3 = len(tramo_3)
total_siniestros = len(data_25mayo)

print(f"Siniestros en tramo PK 0-3000: {count_tramo_1} ({(count_tramo_1/total_siniestros)*100:.2f}%)")
print(f"Siniestros en tramo PK 3000-6000: {count_tramo_2} ({(count_tramo_2/total_siniestros)*100:.2f}%)")
print(f"Siniestros en tramo PK 6000-10000: {count_tramo_3} ({(count_tramo_3/total_siniestros)*100:.2f}%)")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.hist(data_25mayo['PK'], bins=20, edgecolor='black')
plt.title('Distribución de Siniestros por PK en AU 25 de Mayo')
plt.xlabel('Punto Kilométrico (PK)')
plt.ylabel('Cantidad de Siniestros')
plt.axvspan(0, 3000, color='red', alpha=0.3, label='Tramo Crítico 1 (PK 0-3000)')
plt.axvspan(3000, 6000, color='blue', alpha=0.3, label='Tramo Crítico 2 (PK 3000-6000)')
plt.axvspan(6000, 10000, color='green', alpha=0.3, label='Tramo Crítico 3 (PK 6000-10000)')
plt.legend()
plt.show()
```

Test no parametrico **Chi-Cuadrado**

```python
from scipy.stats import chisquare

# Conteos observados
observados = [count_tramo_1, count_tramo_2, count_tramo_3]

# Test Chi-cuadrado de bondad de ajuste
chi2, p = chisquare(f_obs=observados)

print(f"Chi-cuadrado = {chi2:.4f}, p-valor = {p:.4f}")

if p < 0.05:
    print("Rechazamos la hipótesis nula. Las diferencias entre tramos son estadísticamente significativas.")
else:
    print("No hay diferencias significativas entre los tramos.")
```

# **Hipotesis 2**

## La mayoría de los siniestros ocurren en **hora pico**

Utilizamos la variable ``HORA``.

```python
data_h2 = data_preprocessed.copy()
data_h2["Es hora pico"] = data_h2["HORA"].apply(lambda x: "Hora pico" if (7 <= x <= 11) or (16 <= x <= 20) else "No es hora pico")
```

```python
data_h2['HORA'].plot(kind='hist', bins=24, rwidth=0.8)

plt.axvspan(7, 11, color='red', alpha=0.3)
plt.axvspan(16, 20, color='red', alpha=0.3)

plt.title('Cantidad de siniestros por hora')
plt.xlabel('Hora del día')
plt.ylabel('Cantidad de siniestros')
plt.xticks(range(0, 25))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(data_h2["Es hora pico"].value_counts().index, data_h2["Es hora pico"].value_counts().values, color=['tomato', 'dodgerblue'])
plt.title('Siniestros en hora pico vs no hora pico')
plt.ylabel('Cantidad de siniestros')
plt.show()
```

```python
conteo_por_hora = data_h2.groupby("HORA").size().reset_index(name="siniestros")
conteo_por_hora = conteo_por_hora.merge(
    data_h2[["HORA", "Es hora pico"]].drop_duplicates(),
    on="HORA"
)
siniestros_hora_pico = conteo_por_hora[conteo_por_hora["Es hora pico"]=="Hora pico"]["siniestros"]
siniestros_no_hora_pico = conteo_por_hora[conteo_por_hora["Es hora pico"]=="No es hora pico"]["siniestros"]
```

```python
# homocedasticidad con test de levene
from scipy.stats import levene
stat_levene, p_levene = levene(siniestros_hora_pico, siniestros_no_hora_pico)
print(f'Levene: stat={stat_levene}, p={p_levene}')
if p_levene > 0.05:
    print("Se cumple homocedasticidad")
else:
    print("No se cumple homocedasticidad")
```

```python
from scipy.stats import mannwhitneyu

stat_mw, p_mw = mannwhitneyu(
    siniestros_hora_pico,
    siniestros_no_hora_pico,
    alternative="greater"
)

print(f"Mann–Whitney: stat={stat_mw}, p={p_mw}")

if p_mw < 0.05:
    print("Rechazamos H0: en hora pico hay más siniestros por hora")
else:
    print("No rechazamos H0: no hay evidencia de que en hora pico haya más siniestros por hora")
```

# **Hipotesis 4**
## En los siniestros con motos involucradas tiende a haber más heridos que con otro tipo de vehículos.

Utilizamos las variables ``MOTO`` y ``LESIONADOS``

Dividimos los datos en dos grupos: moto y no moto

```python
import matplotlib.pyplot as plt

grupo_moto = data_preprocessed.loc[data_preprocessed['MOTO'] > 0, 'LESIONADOS']
grupo_no_moto = data_preprocessed.loc[data_preprocessed['MOTO'] == 0, 'LESIONADOS']

total_moto = grupo_moto.sum()
total_no_moto = grupo_no_moto.sum()
# imprimir resultados
print(f"Total de lesionados en siniestros con motos: {total_moto}")
print(f"Total de lesionados en siniestros sin motos: {total_no_moto}")
```

Miramos este resultado en un grafico de barras

```python
plt.figure(figsize=(6,5))
plt.bar(['Moto', 'No Moto'], [total_moto, total_no_moto], color=['#1f77b4', '#ff7f0e'])

plt.title('Cantidad total de lesionados por tipo de siniestro')
plt.ylabel('Número total de lesionados')
plt.xlabel('Grupo')

plt.tight_layout()
plt.show()
```

Realizaremos un test-u para comprobar numericamente esta diferencia. Para ello chequeamos previamente sus supuestos

```python
from scipy.stats import levene
# Prueba de Levene para igualdad de varianzas
stat_levene, p_levene = levene(grupo_moto, grupo_no_moto)
print("Prueba de Levene para igualdad de varianzas:")
print("Estadístico:", stat_levene, " | p-valor:", p_levene)
if p_levene < 0.05:
    print("Las varianzas no son iguales.")
else:
    print("Las varianzas son iguales.")
```

```python
from scipy.stats import mannwhitneyu
stat_mw, p_mw = mannwhitneyu(grupo_moto, grupo_no_moto, alternative='greater')
print("\nPrueba de Mann-Whitney U:")
print("Estadístico U:", stat_mw, " | p-valor:", p_mw)
if p_mw < 0.05:
    print("Hay evidencia de que los siniestros con motos tienen más lesionados.")
else:
    print("No hay suficiente evidencia para afirmar que los siniestros con motos tienen significativamente más lesionados.")
```

# **Hipotesis 5**
## La frecuencia de los siniestros viales varía significativamente según la hora del día y el día de la semana

Utilizamos las variables ``FECHA``, ``HORA`` y ``AUTOPISTA``

```python
data_h5 = data_preprocessed.copy()
data_h5['DIA_SEMANA'] = pd.to_datetime(data_h5['FECHA']).dt.day_name()
dias_semana = { 'Monday': 'Lunes', 'Tuesday': 'Martes', 'Wednesday': 'Miércoles',
                 'Thursday': 'Jueves', 'Friday': 'Viernes', 'Saturday': 'Sábado', 'Sunday': 'Domingo' }
data_h5['DIA_SEMANA'] = data_h5['DIA_SEMANA'].replace(dias_semana)

def franja_horaria(h):
    if 6 <= h < 12:
        return 'Mañana'
    elif 12 <= h < 20:
        return 'Tarde'
    elif 20 <= h < 24 or h == 0:
        return 'Noche'
    else:
        return 'Madrugada'

data_h5['FRANJA_HORARIA'] = data_h5['HORA'].apply(franja_horaria)

# Frecuencia de siniestros por franja horaria y día de la semana
tabla = data_h5.groupby(['DIA_SEMANA', 'FRANJA_HORARIA']).size().reset_index(name='CANT_SINIESTROS')
tabla
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Ordenamos para la visualizacion porque si no no se entiende nada
dias_orden = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
franjas_orden = ['Madrugada', 'Mañana', 'Tarde', 'Noche']
tabla['DIA_SEMANA'] = pd.Categorical(tabla['DIA_SEMANA'], categories=dias_orden, ordered=True)
tabla['FRANJA_HORARIA'] = pd.Categorical(tabla['FRANJA_HORARIA'], categories=franjas_orden, ordered=True)
plt.figure(figsize=(10,6))
sns.heatmap(tabla.pivot(index='FRANJA_HORARIA', columns='DIA_SEMANA', values='CANT_SINIESTROS'),
            cmap='YlOrRd', annot=True, fmt='.0f')
plt.title('Frecuencia de siniestros por día y franja horaria')
plt.xlabel('Día de la semana')
plt.ylabel('Franja horaria')
plt.show()
```

```python
autopistas = data_h5[data_h5['AUTOPISTA'].isin([
    'AU 25 DE MAYO', 'AU DELLEPIANE', 'AU PERITO MORENO', 'AU ILLIA', 'AV. LUGONES'])]['AUTOPISTA'].unique()
tabla = data_h5.groupby(['AUTOPISTA', 'DIA_SEMANA', 'FRANJA_HORARIA']).size().reset_index(name='CANT_SINIESTROS')
# heatmap de las autopistas seleccionadas
plt.figure(figsize=(20, 12))
for i, autop in enumerate(autopistas, 1):
    tabla_autop = tabla[tabla['AUTOPISTA'] == autop]
    plt.subplot(3, 3, i)
    #ordenar la tabla para que se vea bien
    tabla_autop['DIA_SEMANA'] = pd.Categorical(tabla_autop['DIA_SEMANA'], categories=dias_orden, ordered=True)
    tabla_autop['FRANJA_HORARIA'] = pd.Categorical(tabla_autop['FRANJA_HORARIA'], categories=franjas_orden, ordered=True)
    
    heat_data = tabla_autop.pivot(index='FRANJA_HORARIA', columns='DIA_SEMANA', values='CANT_SINIESTROS')
    sns.heatmap(heat_data, cmap='YlOrRd', annot=True, fmt='.0f')
    plt.title(f'Autopista: {autop}')
    plt.xlabel('')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#siniestros en lugones miercoles por la mañana
data_h5[(data_h5['AUTOPISTA'] == 'AV. LUGONES') & (data_h5['DIA_SEMANA'] == 'Miércoles') & (data_h5['FRANJA_HORARIA'] == 'Mañana')]
```

Test Chi-cuadrado --> Queremos saber si la frecuencia de siniestros cambia según el día de la semana o franja horaria

```python
from scipy.stats import chi2_contingency

# Tabla de contingencia: filas = día, columnas = franja
tabla = pd.crosstab(data_h5['DIA_SEMANA'], data_h5['FRANJA_HORARIA'])

# Test Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)
print(f"Chi2 = {chi2:.2f}, p-valor = {p:.4f}")
if p < 0.05:
    print("Hay una relación significativa entre el día de la semana y la franja horaria en la frecuencia de siniestros.")
else:
    print("No hay evidencia de una relación significativa entre el día de la semana y la franja horaria en la frecuencia de siniestros.")
```

# **Hipotesis 6**
## La AU Perito Moreno presenta un riesgo significativamente mayor de colisiones con obstáculos fijos

Usamos las variables `AUTOPISTA` y ``TIPO DE SINIESTRO``

```python

data_perito_moreno = data_preprocessed[(data_preprocessed["AUTOPISTA"] == "AU PERITO MORENO") & (data_preprocessed["TIPO DE SINIESTRO"] == "COLISION CON OBSTACULO FIJO")]
autopistas_principales = data_preprocessed[data_preprocessed["TIPO DE SINIESTRO"] == "COLISION CON OBSTACULO FIJO"]["AUTOPISTA"].value_counts().head(5).index.tolist()

siniestros_obstaculo_fijo_counts = {autopista: data_preprocessed[(data_preprocessed["AUTOPISTA"] == autopista) & (data_preprocessed["TIPO DE SINIESTRO"] == "COLISION CON OBSTACULO FIJO")].shape[0] for autopista in autopistas_principales}
plt.bar(siniestros_obstaculo_fijo_counts.keys(), siniestros_obstaculo_fijo_counts.values(), color='firebrick')
plt.xlabel('Autopista')
plt.ylabel('Cantidad de siniestros')
plt.title('Cantidad de siniestros con obstáculo fijo por autopista')
plt.xticks(rotation=45)
plt.show()
```

```python
indice_riesgo = {}
for autopista in autopistas_principales:
    total_siniestros = data_preprocessed[data_preprocessed["AUTOPISTA"] == autopista].shape[0]
    siniestros_obstaculo_fijo = data_preprocessed[(data_preprocessed["AUTOPISTA"] == autopista) & (data_preprocessed["TIPO DE SINIESTRO"] == "COLISION CON OBSTACULO FIJO")].shape[0]
    indice_riesgo[autopista] = siniestros_obstaculo_fijo / total_siniestros 
indice_riesgo
indice_riesgo = pd.Series(indice_riesgo).sort_values(ascending=False)
```

```python
plt.bar(indice_riesgo.index, indice_riesgo.values, color='tomato')
plt.xlabel('Autopista')
plt.ylabel('Índice de riesgo')
plt.title('Índice de colisión con obstáculo fijo por autopista')
plt.xticks(rotation=45)
plt.show()
```
